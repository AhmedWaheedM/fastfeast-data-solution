```python
import logging
import structlog
from logging.handlers import TimedRotatingFileHandler
```

```python
end_time = datetime.now(timezone.utc)
```

```python
log_file = "channels.log"
rotation_handler = TimedRotatingFileHandler(
    log_file,
    when="midnight",
    interval=1,
    backupCount=7
)
```

```python
logging.basicConfig(
    level=logging.INFO,
    format="%(message)s",
    handlers=[rotation_handler],
    force=True
)
```

```python
structlog.configure(
    processors=[
        structlog.stdlib.add_log_level,
        structlog.processors.dict_tracebacks,
        structlog.processors.TimeStamper(fmt="iso", key="log_timestamp"),
        structlog.processors.JSONRenderer()
    ],
    logger_factory=structlog.stdlib.LoggerFactory(),
    wrapper_class=structlog.stdlib.BoundLogger,
    cache_logger_on_first_use=True,
)
logger = structlog.get_logger()
```

```python
logger.info(
       'Ingestion Run',
        pipeline_context={
            "ingestion_type": "batch",
            "file_name": file_name,
            "total_records_processed": '',
            "processing_start_time": start_time.isoformat(),
            "processing_end_time": end_time.isoformat(),
        },
        schema_validation={
            "is_correct": '',
            "expected_column_count": len(columns),
            "actual_column_count": '',
        },
        column_metrics=[
            {
                "column_name": columns[0]['name'],
                "data_type": {
                    "expected_column_data_type": columns[0]['type'],
                    "is_valid_column_data_type": '',
                    "count_invalid_column_data_type":'',
                    "count_valid_column_data_type":'',
                },
                "null_validation": {
                    "is_nullable": columns[0]['nullable'],
                    "null_count": 0
                },
                "constraint_violations": {
                    "primary_key_violations_count": '',
                }
            },
            {
                "column_name": columns[1]['name'],
                "null_validation": {
                    "is_nullable": columns[1]['nullable'],
                    "null_count": ''
                },
                "range_validation": {
                    "expected_min": 18,
                    "expected_max": 120,
                    "in_range_count": '',
                    "out_of_range_count": ''
                }
            }
        ]
    )
```
