```python
import logging
import queue
import structlog
from logging.handlers import TimedRotatingFileHandler, QueueHandler, QueueListener
```

```python
log_file = "channels.log"
rotation_handler = TimedRotatingFileHandler(
    log_file,
    when="midnight",
    backupCount=7
)
```

# **Creating a queue & its handler** 🧮

```python
log_queue = queue.Queue(-1)
queue_handler = QueueHandler(log_queue)
```

# **Configuring the logger to use the queue "non-blocking" :)**

```python
logging.basicConfig(
    level=logging.INFO,
    handlers=[queue_handler],
    force=True
)
```

# **Releasing the listener on the queue** 🦻

```python
listener = QueueListener(log_queue, rotation_handler)
listener.start()
```

# **Handing over to the structlog** 🕵️‍♂️

```python
structlog.configure(
    processors=[
        structlog.stdlib.add_log_level,
        structlog.processors.TimeStamper(fmt="iso", key="log_timestamp"),
        structlog.processors.JSONRenderer()
    ],
    logger_factory=structlog.stdlib.LoggerFactory(),
    wrapper_class=structlog.stdlib.BoundLogger,
)

logger = structlog.get_logger()
```

```python
logger.info("Happy non-blocking logging!!", pipeline_context={"EXP3": "SO MUCH FUN"})
```
