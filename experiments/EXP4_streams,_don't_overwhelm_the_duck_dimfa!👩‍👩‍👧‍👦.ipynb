# **Collecting LEGOs 𐦂𖨆𐀪𖠋**

```python
import multiprocessing
import time
import random
```

# **Planning the crime 🦖**

```python
def criminate_then_escape(criminal, status_dict, slang):
  print(f"From criminal {criminal} to the gang members: ")
  print(f"We will be having so much fun, HAHAHAHHA!")
  print(f"over\n")
  print(f"ME {criminal}, Dn 👔")

  if criminal != 'orders':
    if criminal == 'tickets_events':
      while status_dict.get('orders') != slang:
        time.sleep(0.47)
    if criminal == 'tickets':
      while status_dict.get('tickets_events') != slang:
        time.sleep(0.47)

  print(f'I am "{criminal}" escaping, you can\'t catch me!👻')
  status_dict[criminal] = slang
```

# **Committing it! 🤜🤛**

```python
if __name__ == "__main__":
  gang_leader = multiprocessing.Manager()
  shared_status = gang_leader.dict()

  criminals = ['orders', 'tickets_events', 'tickets']
  processes = []

  for criminal in criminals:
    p = multiprocessing.Process(target=criminate_then_escape, args=(criminal, shared_status, "MOFASA"))
    processes.append(p)
    p.start()

  for p in processes:
        p.join()
```
